```python
#question 3
from scipy import stats
group_a=[110,115,108,102,107]
group_b=[120,125,130,122,128]
```

```python
t_stat,p_value=stats.ttest_ind(group_a,group_b)
```

```python
print(group_a)
print(group_b)
print(f"\nT-statistic:{t_stat:.2f}")
print(f"P-value:{p_value:.4f}")
```

```python
#Question 4
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data={'Department':['HR','IT','Sales','HR','IT',np.nan,'Sales','HR','IT','Sales'],'Experience_Years':[1,0,3,2,np.nan,4,5,3,2,1],'Projects_completed':[2,0,5,3,4,np.nan,6,2,3,4],'Performance_Rating':[3,0,2,4,3,5,np.nan,3,4,2]}
df=pd.DataFrame(data)
print(df)
```

```python
df_clean=df.copy()
df_clean['Department'].fillna('Unknown',inplace=True)
df_clean['Experience_Years'].fillna(df_clean['Experience_Years'].median(),inplace=True)
df_clean['Projects_completed'].fillna(df_clean['Projects_completed'].median(),inplace=True)
df_clean['Performance_Rating'].fillna(df_clean['Performance_Rating'].mean(),inplace=True)
print(df_clean)
```

```python
print(df_clean.describe())
```

```python
corr=df_clean[['Experience_Years','Projects_completed,'Performance_Rating']].corr()
print(corr)
sns.heatmap(corr,annot=True)
```

```python
sns.hisplot(df_clean)
sns.boxplot(data=df_clean)
sns.scatterplot(data=df_clean)
```
